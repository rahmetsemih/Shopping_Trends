```python
gender_data=data[['Subscription Status','Discount Applied','Promo Code Used','Gender','Previous Purchases']]
gender_data
```

```python
X=gender_data.drop(['Gender'],axis=1)
Y=gender_data['Gender']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
```

```python
 model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared  Error:", mse)
```

```python

plt.figure(figsize=(10, 8))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.title('Gerçek Değerler vs. Tahmin Edilen Değerler')
plt.xlabel('Gerçek Değerler')
plt.ylabel('Tahmin Edilen Değerler')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], '--k', linewidth=2)
plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
X_1, y_1 = gender_data.drop("Gender", axis=1), yeni_data['Gender']
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X_1, y_1, test_size=0.15)
```

```python
forest = RandomForestRegressor(random_state=42)
forest.fit(X_train_1, y_train_1)
```

```python
y_pred_1 = forest.predict(X_test_1)
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(y_test_1, y_pred_1, alpha=0.5)
plt.plot([y_test_1.min(), y_test_1.max()], [y_test_1.min(), y_test_1.max()], '--k', linewidth=2, color="red")
plt.title('Gerçek Değerler vs. Tahmin Edilen Değerler')
plt.xlabel('Gerçek Değerler')
plt.ylabel('Tahmin Edilen Değerler')

plt.show()
```

```python
forest.score(X_test_1,y_test_1)
```
