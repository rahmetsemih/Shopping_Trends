## Veri setimizde birbirini tekrar eden kolon olmadığını göstermiş olduk.


```python
plt.subplots(figsize=(10,10))
sns.histplot(data['Age'], bins=15, kde=True)
plt.title('Age Distribution')

```

```python
plt.figure(figsize = (20, 6))
xy = data["Gender"].value_counts().plot(kind = 'bar',  rot = 10)
xy.set_xticklabels(('Male', 'Female'))
```

```python
plt.figure(figsize = (20, 8))

counts = data["Gender"].value_counts()
explode = (0, 0.01)

counts.plot(kind = 'pie', fontsize = 15,  explode = explode, autopct = '%1.f%%')
plt.xlabel('Gender', weight = "bold", color = "#0f475d", fontsize = 14, labelpad = 20)
plt.axis('equal')
plt.legend(labels = counts.index, loc = "best")
plt.show()
```

```python
fig=px.box(data,x='Purchase Amount (USD)',title='USD',color_discrete_sequence=["#7A1012"])
#SHOW THE PLOT
fig.show()
```

```python
plt.hist(data["Review Rating"], color=["#7A1012"], edgecolor='black')
plt.xlabel('Review Rating')
plt.title('Review Rating Distrubition') 
plt.show()
```

```python
sns.countplot(x=data["Size"],data=data)
plt.ylabel("Count")
plt.title("Size")
plt.show()
```

```python
fig = px.pie(data, names='Size', values='Customer ID', title='Ratio of Size')
fig.show()
```
